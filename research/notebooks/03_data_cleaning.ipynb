# Data Cleaning & Normalization

Filters outliers and currency symbols.


## Data Cleaning, Deduplication, and Normalization


```python
import pandas as pd
import numpy as np
import os

df = pd.read_csv("research/datasets/raw/real_estate_leads.csv")

# 1. Duplicate removal
df = df.drop_duplicates()

# 2. Currency normalization
df["price_usd"] = df["price_raw"].str.replace("$", "").str.replace(",", "").astype(float)

# 3. Handle outliers using Z-score or IQR
sqft_mean = df["square_feet"].mean()
sqft_std = df["square_feet"].std()
# Filter values within 3 standard deviations
df = df[(df["square_feet"] >= sqft_mean - 3 * sqft_std) & (df["square_feet"] <= sqft_mean + 3 * sqft_std)]

# Save clean dataset
os.makedirs("research/datasets/processed", exist_ok=True)
df.to_csv("research/datasets/processed/real_estate_leads_clean.csv", index=False)
print(f"Cleaned dataset saved. Remaining samples: {len(df)}")
```
