Ingests raw telemetry files.


## Ingesting and Saving Raw Real Estate Marketing Telemetry


```python
import os
import pandas as pd
import numpy as np

# Generate synthetic raw real estate listings & leads telemetry dataset
np.random.seed(42)
n_samples = 1500

data = {
    "listing_id": [f"lst_{1000 + i}" for i in range(n_samples)],
    "price_raw": [f"${int(np.random.normal(350000, 100000))}" for _ in range(n_samples)],
    "square_feet": np.random.normal(2000, 500, size=n_samples).astype(int),
    "bedrooms": np.random.choice([1, 2, 3, 4, 5], size=n_samples),
    "bathrooms": np.random.choice([1.0, 1.5, 2.0, 2.5, 3.0], size=n_samples),
    "ad_spend_usd": np.random.uniform(50, 1500, size=n_samples),
    "impressions": np.random.randint(100, 20000, size=n_samples),
    "clicks": np.random.randint(5, 500, size=n_samples),
    "leads_converted": np.random.choice([0, 1], size=n_samples, p=[0.75, 0.25]),
    "description": np.random.choice([
        "Beautiful family house in the suburbs with a large green garden.",
        "Modern downtown condo with state-of-the-art kitchen features.",
        "Cozy cottage nestled in the woods near a quiet lake.",
        "Luxury penthouse overlooking the city skyline with a private balcony."
    ], size=n_samples)
}

# Inject duplicate listing IDs to simulate raw telemetry errors
df = pd.DataFrame(data)
df = pd.concat([df, df.iloc[:15]], ignore_index=True)

# Save raw dataset
os.makedirs("research/datasets/raw", exist_ok=True)
df.to_csv("research/datasets/raw/real_estate_leads.csv", index=False)
print("Raw real estate leads dataset saved successfully.")
```
