# Exploratory Data Analysis

Summarizes metrics and distributions.


## Exploratory Data Analysis Dashboard


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv("research/datasets/processed/real_estate_leads_features.csv")

# Print dataset summary statistics
print(df.describe())

# 1. Target conversion distribution
print("\nConversion (leads_converted) Distribution:")
print(df["leads_converted"].value_counts(normalize=True))

# 2. Correlation of numerical features
numeric_df = df.select_dtypes(include=[np.number])
print("\nFeature Correlation Matrix:")
print(numeric_df.corr())

# 3. Simple histogram visualization
plt.figure(figsize=(8, 4))
plt.hist(df["price_usd"], bins=30, edgecolor='black')
plt.title("Distribution of Real Estate Prices")
plt.xlabel("Price in USD")
plt.ylabel("Count")
plt.show()
```
