# Schema & Bound Validation

Checks column types and bounds.


## Validating Raw Dataset Ingestion


```python
import pandas as pd
import os

df = pd.read_csv("research/datasets/raw/real_estate_leads.csv")

# 1. Assert required columns are present
expected_cols = ["listing_id", "price_raw", "square_feet", "bedrooms", "ad_spend_usd", "impressions", "clicks", "leads_converted"]
for col in expected_cols:
    assert col in df.columns, f"Missing required column: {col}"

# 2. Check for missing value counts
missing_counts = df.isnull().sum()
print("Validation Report: Missing Values count:")
print(missing_counts)

# 3. Assert value bounds
assert (df["square_feet"] > 0).all(), "Invalid square_feet found!"
assert (df["ad_spend_usd"] >= 0).all(), "Invalid negative ad spend values found!"

print("Dataset passed schema validation successfully.")
```
