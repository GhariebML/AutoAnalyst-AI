# Feature Engineering Pipeline

Vectorizes text and defines KPIs.


## Feature Engineering & Vectorization Pipeline


```python
import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import TfidfVectorizer
import os

df = pd.read_csv("research/datasets/processed/real_estate_leads_clean.csv")

# 1. Text feature TF-IDF vectorizer
vectorizer = TfidfVectorizer(max_features=5)
tfidf_feats = vectorizer.fit_transform(df["description"]).toarray()
tfidf_df = pd.DataFrame(tfidf_feats, columns=[f"tfidf_{i}" for i in range(5)])

# 2. Marketing KPI Engineering (CTR & Cost-Per-Click)
df["ctr"] = df["clicks"] / df["impressions"]
df["cpc"] = df["ad_spend_usd"] / (df["clicks"] + 1)

# Concat features
processed_df = pd.concat([df.reset_index(drop=True), tfidf_df.reset_index(drop=True)], axis=1)

# Save processed dataset
processed_df.to_csv("research/datasets/processed/real_estate_leads_features.csv", index=False)
print("Features generated and saved to processed folder.")
```
