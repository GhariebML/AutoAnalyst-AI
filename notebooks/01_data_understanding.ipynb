2. DATASET OVERVIEW & BRIEF DESCRIPTION
--------------------------------------------------------------------------------
The Credit Scoring Dataset evaluates the creditworthiness of individuals based
on demographic, financial, and behavioral attributes. It helps financial
institutions mitigate risks and make data-driven lending decisions.

* Data Source: https://www.kaggle.com/datasets/laotse/credit-risk-dataset?resource=download
* Total Instances (Rows): 32,581
* Total Attributes (Columns): 12 (11 Features + 1 Target)


3. DATA DICTIONARY
--------------------------------------------------------------------------------
Below is the definition and structural role of each column in the dataset:

- person_age                  [Integer]     : Age of the loan applicant.
- person_income               [Integer/Float]: Annual income of the applicant.
- person_home_ownership       [Categorical] : Home status (RENT, OWN, MORTGAGE, OTHER).
- person_emp_length           [Float]       : Length of employment in years.
- loan_intent                 [Categorical] : Purpose of the loan (PERSONAL, MEDICAL, etc).
- loan_grade                  [Categorical] : Credit grade assigned to the loan (A–G).
- loan_amnt                   [Integer/Float]: Total amount of the loan requested.
- loan_int_rate               [Float]       : Interest rate applied to the loan.
- loan_status                 [Binary (0/1)]: TARGET VARIABLE (1 = default, 0 = fully paid).
- loan_percent_income         [Float]       : Ratio of loan amount to annual income.
- cb_person_default_on_file   [Categorical] : Historical record of prior default (Y/N).
- cb_person_cred_hist_length  [Integer]     : Length of applicant's credit history (years).


4. BASIC DATA PROFILING & QUALITY ASSESSMENT

A. MISSING VALUES (NULLS):
   - ' person_emp_length ': Has missing entries (needs imputation or handling).
   - ' loan_int_rate ': Has missing interest rates (requires treatment before ML).

   - Any identical rows will be dropped to prevent bias.

C. PRE-PROCESSING OBSERVATIONS:

5. PROPOSED MACHINE LEARNING MODELS
--------------------------------------------------------------------------------
Since the target variable 'loan_status' is Binary (0 or 1), this is a
Binary Classification problem. We recommend evaluating the following models:

1. Logistic Regression:
   - A great baseline model for financial risk, easy to interpret and explain
     to stakeholders.
2. Random Forest Classifier:
   - Excellent for handling non-linear relationships and interactions
     between features (like income vs. loan amount).
3. XGBoost / LightGBM:
   - Advanced boosting algorithms that usually give the highest predictive
